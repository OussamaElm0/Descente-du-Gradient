# f(a) = (a + 5)²
 - Learning rate = 0.01
 - Start point = -3

```python
def derive(a):
    return 2 * (a + 5)

a = -3
alpha = 0.01
old_a = a
```

## 1000 itérations

```python
n_iterations = 1000

for i in range(n_iterations):
    a = a - alpha * derive(a)

print(a)
```

## 100000 itérations

```python
n_iterations = 100000

for i in range(n_iterations):
    a = a - alpha * derive(a)
    print(f"Iteration : {i} Valeur de a : {a}")
    if old_a == a :
        break
    old_a = a

print(a)
```

## 100000000 itérations

```python
n_iterations = 100000000

for i in range(n_iterations):
    a = a - alpha * derive(a)
    print(f"Iteration : {i} Valeur de a : {a}")
    if old_a == a :
        break
    old_a = a

print(a)
```
