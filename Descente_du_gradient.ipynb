# f(a) = (a + 5)²
 - Learning rate = 0.01
 - Start point = -3

```python
def derive(a):
    return 2 * (a + 5)

a = -3
alpha = 0.01
old_a = a
```

## 1000 itérations

```python
n_iterations = 1000

for i in range(n_iterations):
    a = a - alpha * derive(a)

print(a)
```

## 100000 itérations

```python
n_iterations = 100000

for i in range(n_iterations):
    a = a - alpha * derive(a)
    print(f"Iteration : {i} Valeur de a : {a}")
    if old_a == a :
        break
    old_a = a

print(a)
```

## 100000000 itérations

```python
n_iterations = 100000000

for i in range(n_iterations):
    a = a - alpha * derive(a)
    print(f"Iteration : {i} Valeur de a : {a}")
    if old_a == a :
        break
    old_a = a

print(a)
```

# f(x)=x²+4x+4
 - Learning rate = 0.01
 - Start point = 10

```python
def derive(x):
    return (2 * x) + 4

alpha = 0.01
x = 10
old_x = x
```

## 1000 itérations

```python
for _ in range(1000):
    x -= alpha * derive(x)
    print(f"Iteration: {_} Valeur de x : {x}")
    if old_x == x:
        break
    old_x = x

print(x)
```

## 10000 itérations


```python
for _ in range(10000):
    x -= alpha * derive(x)
    print(f"Itération: {_} Valeur de x: {x}")
    if old_x == x:
        break

    old_x = x

print(x)
```
